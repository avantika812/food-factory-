```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
url="https://en.wikipedia.org/wiki/Food_Factory"
```

```python
page=requests.get(url,verify=False)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find('table',{"class":"wikitable"})
```

```python
table=table.find("tbody")
table
```

```python
rows=table.find_all('tr')
```

```python
rows
```

```python
info=[]
for i in rows:
    cols=i.find_all('td')
    #print(cols)
    food_info=[c.text for c in cols]
    #print(food_info)
    info.append(food_info)
print(info) 
```

```python
info[1]
```

```python
Episode = []
Title = []
Segment_A = []
Segment_B = []
Segment_C = []
Segment_D = []

a=[]
for i in info:
    a.append(i)
```

```python
b=a[1:]
```

```python
for i in b:
    Episode.append(i[0])
    Title.append(i[1])
    Segment_A.append(i[2])
    Segment_B.append(i[3])
    Segment_C.append(i[4])
    Segment_D.append(i[5])
```

```python
Episode
```

```python
Data={'Episode':Episode,'Title':Title,'Segment A':Segment_A,'Segment B':Segment_B,'Segment C':Segment_C,'Segment D': Segment_D}
```

```python
food_info=pd.DataFrame(Data)
food_info
```

```python
food_info.to_csv('food.csv')
```
